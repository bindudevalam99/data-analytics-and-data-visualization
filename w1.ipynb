```python
#importing libraries
import numpy as np
import pandas as pd
```

```python
# reading dataset
df = pd.read_csv('weather_dataset.csv')
df.head(10)
```

# Subtask 1: Correct years for given data set

```python
df.set_index('Date').sort_index()
```

```python
# removing the special character
df.replace('\'','', regex=True, inplace=True)
```

```python
# removing the special character
df.replace(';','', regex=True, inplace=True)
```

# Subtask 2: removal of duplicate rows and duplicate Columns

```python
#removal of duplicate rows and duplicate Columns
df.duplicated()
```

```python
df1=df.drop_duplicates()
df1
```

```python
#drops the unnamed columns
df1.drop(df1.columns[df1.columns.str.contains('unnamed',case = False)],axis = 1, inplace = True)
df1
```

```python
#drop the Nan rows
df1.dropna(axis=0,inplace=True)
df1
```

```python
# Removing the date 2022-02-29 because there is no 29th in february in leap year
df1 = df1[df1.Date != '2022-02-29']
df1
```

# Subtask 3: fix a few labels in the given data set 

```python
#finding the datatypes
df1.dtypes
```

```python
# converting all to float except date, date1
cols=[' Temperature', ' Average humidity (%', ' Average dewpoint (°F',
       ' Average barometer (in', ' Average windspeed (mph',
       ' Average gustspeed (mph', ' Average direction (°deg',
       ' Rainfall for month (in', 'Rainfall for year (in',
       ' Maximum rain per minute', ' Maximum temperature (°F',
       ' Minimum temperature (°F', ' Maximum humidity (%',
       ' Minimum humidity (%', ' Maximum pressure', ' Minimum pressure',
       ' Maximum windspeed (mph', ' Maximum gust speed (mph',
       ' Maximum heat index (°F', ' diff_pressure']
df1[cols]=df1[cols].apply(pd.to_numeric,errors='coerce',axis=1) 
df1.dtypes
```

```python
#converting month to int
df1[' Month'] = df1[' Month'].astype(int)
df1.dtypes
```

```python
# renaming the columns properly
df1.set_axis(['Date', 'Temperature', 'Average humidity (%)', 'Average dewpoint (°F)',
       'Average barometer (in)', 'Average windspeed (mph)',
       'Average gustspeed (mph)', 'Average direction (°deg)',
       'Rainfall for month (in)', 'Rainfall for year (in)', 'Maximum rain per minute',
           'Maximum temperature (°F)', 'Minimum temperature (°F)',
           'Maximum humidity (%)' , 'Minimum humidity (%)', 'Maximum pressure','Minimum pressure',
           'Maximum windspeed (mph)', 'Maximum gust speed (mph)','Maximum heat index(°F)', 'Date1',
           'Month','diff_pressure'], axis='columns', inplace=True)
```

```python
df1.dtypes
```

```python
df1.to_csv('my_result.csv', encoding='utf-8', index=False)
```
